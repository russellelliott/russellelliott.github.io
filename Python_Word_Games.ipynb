```
#@title Generate the list of words
#importing the time module
import time
time.sleep(0.5)
print("Generating word list...")

import random
#here we set the secret word
#options = ["secret", "bishop", "canyon", "salamander"]
options = []
#Secret word is a random word from a long list of words

import urllib.request
urllist = []

#Check from multiple txt files, as some txt files have a word that the others don't.

#These txt files have all the words in lower case
#urllist.append("http://www.russellelliott.codehs.me")
urllist.append("http://www.math.sjsu.edu/~foster/dictionary.txt")
#58,110 words
urllist.append("http://www.mieliestronk.com/corncob_lowercase.txt")
urllist.append("https://www.mit.edu/~ecprice/wordlist.10000")

#This txt file has all the words in upper case
#47,162 words
urllist.append("http://programarcadegames.com/python_examples/en/dictionary.txt")

#This file has numbers in it
#7776 words
urllist.append("https://www.eff.org/files/2016/07/18/eff_large_wordlist.txt")

#This txt files have dashes and spaces, which are removed by only including characters in the string.
#634,699 words
urllist.append("https://wordlists.capsop.com/english.txt")
#urllist.append("https://wordlists.capsop.com/webster-dictionary.txt")
urllist.append("https://raw.githubusercontent.com/dwyl/english-words/master/words_alpha.txt")

for i in range(len(urllist)):
    url = urllist[i]
    #file = urllib.request.urlopen(url)
    req = urllib.request.Request(url, headers={"User-Agent": "Chrome"})
    file = urllib.request.urlopen(req)
    
    for line in file:
    	#decoded_line = line.decode("utf-8")
    	#Modified unicode ID so it works for all the txt files.pro
    	#https://stackoverflow.com/questions/19699367/for-line-in-results-in-unicodedecodeerror-utf-8-codec-cant-decode-byte
    	decoded_line = line.decode("ISO-8859-1")
    	#Makes the line lowercase in case its not.
    	decoded_line = decoded_line.lower()
    	#removes spaces that may appear
    	#https://www.journaldev.com/23763/python-remove-spaces-from-string
    	decoded_line = decoded_line.strip()
    	
    	decoded_line = decoded_line.replace("'", "")
    	
    	#https://www.geeksforgeeks.org/python-ways-to-remove-numeric-digits-from-given-string/
    	# using join and isdigit  
    	# to remove numeric digits from string
    	#The fourth txt file has numbers next to the words. This function removes the numbers, prohibiting the user
    	#from entering a number as a word.
    	newline = ''.join([i for i in decoded_line if i.isalpha()])
    	#if not i.isdigit()
    	decoded_line = newline
    	#adds word to the list
    	#result = options.index(decoded_line)
    	#if result!=-1:
    	if len(decoded_line)>=3:
    	    #and decoded_line not in options
    	    # not in options
    	    #Only adds the word if it is greater than or equal to 3 characters long and if it's not in the list.
    	    #Words like a, i, or be won't be added, as they are too small for a meaningful game.
    	    options.append(str(decoded_line))

#https://www.w3schools.com/python/python_howto_remove_duplicates.asp
options = list(dict.fromkeys(options))
print("The word list is generated. ready to play!")
```

<h1>Python Hangman</h1>
<p>This is Hangman coded in Python. Click the play button to run the program. You will have 10 guesses to guess a random word.</p>

```
#@title Hangman Game Execute (run this cell to play Hangman)
print("HANGMAN")
print("Start guessing!")
length = len(options)
word = options[random.randint(0,len(options)-1)]

#creates an variable with an empty value
guesses = ''

#letters already guessed, but failed
guessed = []
#determine the number of turns
#number of turns/chances depends on length of the word?

#turns is number of chances. A paly can solve the puzzle without wrong answers, effectively having unlimited chances.
turns = 10

letters = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m", "n", "o", "p", "q", "r", "s", "t", "u", "v", "w", "x", "y", "z"]

# Create a while loop

#check if the turns are more than zero
while turns > 0:         

    # make a counter that starts with zero
    failed = 0             

    # for every character in secret_word    
    for char in word:      

    # see if the character is in the players guess
        if char in guesses:    
    
        # print then out the character
            print(char, end = " ")    

        else:
    
        # if not found, print a dash
            print("_", end = " ")     
       
        # and increase the failed counter with one
            failed += 1    

    print()
    # if failed is equal to zero
    # print You Won
    if failed == 0:        
        print ("You won!")

    # exit the script
        break              

    #print

    # ask the user go guess a character
    #guess = input("guess a character: ")
    #https://www.geeksforgeeks.org/list-methods-in-python-set-1-in-not-in-len-min-max/
    while True:
        guess = input("Guess a character: ")
        #letters is an array containing all the letters of the English alphabet. Every letter you guess is removed from this array.
        #So, if you attempt to guess that letter again, it won't work, as that letter isn't in the letters array.
        if guess not in letters:
            print("You already guessed that letter!")
        else:
            break

    # set the players guess to guesses
    guesses += guess
    letters.remove(guess)

    # if the guess is not found in the secret word
    if guess not in word:  
 
     # turns counter decreases with 1 (now 9)
        turns -= 1        
 
    # print wrong
        print("Wrong")
        guessed.append(guess)
        #print("Wrong Letters Guessed")
        print(guessed)
 
    # how many turns are left
        print ("You have", + turns, 'more guesses') 
 
    # if the turns are equal to zero
        if turns == 0:           
    
        # print "You Lose"
            print ("You lose!")
            print ("The word was: " + word)
            print ("Better luck next time!")
```
